```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy.special import expit
```

```python
df = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
df
```

```python
#edit_df = df[~df['Country/Region'].str.contains("China")]
edit_df = df[df['Country/Region'].isin(['US'])]
df = edit_df
df
```

```python
df = df.drop(["Province/State", "Country/Region" ,"Lat", "Long"], axis=1)
df
```

```python
death_by_day = list(df.sum(axis=0))
death_by_day
```

```python
def sigmoid(x, a, b, c):
     return (c / (1 + np.exp(-b*(x-a))))
```

```python
x = np.array(death_by_day)
y = np.array(list(range(len(death_by_day))))
popt, pcov = curve_fit(sigmoid, y, x, bounds=([0, 0, death_by_day[-1]], [60, 0.15, 100000]))
print(popt)
```

```python
%matplotlib notebook
extra = 30
plt.plot([sigmoid(i, popt[0], popt[1], popt[2]) for i in range(len(death_by_day) + extra)], color='r')
plt.plot(death_by_day, color='b')
plt.ylabel('Total Deaths')
plt.xlabel('Day Number (since 1/22/2020)')
plt.show()
print("Current Deaths: ", death_by_day[-1])
print("Total predicted deaths: ", sigmoid((len(death_by_day) + 10000), popt[0], popt[1], popt[2]))
print("Sigmoid params: ", popt[0], popt[1], popt[2])
```
