```python
def sigmoid(x, a, b, c):
     return (c / (1 + np.exp(-b*(x-a))))
```

```python
def predict(pred_country):
    '''Predict covid19 for a specified country. Makes fundemental assumption that countries are independent, 
    and that they are near a sigmoid curve. '''
    
    active = {}
    
    # remove all countries with less total days than country day + 1 (not predictive)
    p_df = df[df.iloc[:,-1] > 20]
    p_df = p_df[p_df['Country/Region'].isin([pred_country])]
    country_activity = []
    days_active = len(country_activity)
    p_df = df[df.iloc[:,-1] > 20]
    
    for c in set(df[df.iloc[:,-1] > 20].iloc[:, 0]):
        z = p_df[p_df.iloc[:,-1] > 20]
        z = z[z['Country/Region'].isin([c])]
        activity = [i for i in list(z.sum(axis=0))[1:] if i != 0]
        if (c == pred_country):
            country_activity = activity
        elif (len(activity)) > days_active:
            active[c] = activity
        #else:
        #    p_df.drop(z[z['Country/Region'] == c].index, axis=0, inplace=True)
        
    # normalize
    
    total = np.array([0.,0.,0.])
    # calculate avg sigmoid params
    for country, deaths_by_day in active.items():
        death_by_day = deaths_by_day / pop_by_country[country]
        x = np.array(death_by_day)
        y = np.array(list(range(len(death_by_day))))
        popt, pcov = curve_fit(sigmoid, y, x, bounds=([0, 0, 0], [60, 0.25, 0.000275]))
        total += np.array(popt)
    
    total /= len(active)
    # regression type penalty to try to match avg sigmoid params to curve ? 
    death_by_day = country_activity / pop_by_country[pred_country]
    x = np.array(death_by_day)
    y = np.array(list(range(len(death_by_day))))
    popt, pcov = curve_fit(sigmoid, y, x, bounds=([0, 0, 0], [60, 0.25, 0.000275]))
    %matplotlib notebook
    extra = 30
    plt.plot([sigmoid(i, popt[0], popt[1], popt[2]) for i in range(len(death_by_day) + extra)], color='r')
    plt.plot(death_by_day, color='b')
    plt.ylabel('Total Deaths')
    plt.xlabel('Day Number (since first case)')
    plt.show()
    print("Current Deaths: ", death_by_day[-1])
    print("Total predicted deaths: ", popt[2])
    print("Sigmoid params: ", popt[0], popt[1], popt[2])
    
    # TODO:
    # k cross-validation to optimize lambda in regression
    
    # return sigmoid params
    return total
```
